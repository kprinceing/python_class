#### 例12-1 matplotlib简单绘图

```python
"""
========================
A more complex fill demo
========================

In addition to the basic fill plot, this demo shows a few optional features:

    * Multiple curves with a single command.
    * Setting the fill color.
    * Setting the opacity (alpha value).
"""
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)
y1 = np.sin(x)
y2 = np.sin(3 * x)

fig, ax = plt.subplots()
ax.fill(x, y1, 'b', x, y2, 'r', alpha=0.3)
plt.show()
```

#### 例12-3 使用matplotlib绘制完整图形

```python
import math
import matplotlib.pyplot as plt 

X = [ i / 100  for i in range(0, 1000) ]
Y = [ math.sin(x) for x in X ]
Z = [ math.cos(x**2) for x in X ]

plt.figure(figsize=(8,4))
plt.plot( X , Y , label="$sin(x)$" , color="red" , linewidth=2 ) 
plt.plot( X , Z , "b--" , label="$cos(x^2)$" ) 
plt.xlabel("Time(s)") 
plt.ylabel("Volt") 
plt.title("PyPlot First Example") 
plt.ylim(-1.2,1.2) 
plt.legend() 
plt.show()
```

```python
# 线宽、线型、网格线控制
import matplotlib.pyplot as plt
import math

X = [ i / 100  for i in range(0, 1000) ]
sin=[ math.sin(x) for x in X ]
cos=[ math.cos(x) for x in X ]

plt.plot(X,sin,color='blue',linewidth=2.0, linestyle='-') 
# 设置蓝色、2像素宽划线
plt.plot(X,cos,color='red',linewidth=1.0, linestyle='-.') 
# 用虚线画图
plt.grid(True)
# 画网格线
plt.show()
```

#### 例12-4 中国人口出生率、死亡率、自然出生率变化

```python
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline 
#%matplotlib qt5

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
mpl.rcParams['axes.unicode_minus'] = False # 解决保存图像是负号'-'显示为方块的问题

with open('中国人口出生率年度数据.csv') as f:
    datas = f.readlines()
    title = datas[0].strip().split(',') 
    years = [ int(y[:-1]) for y in title[1:]]
    d1 = datas[1].strip().split(',')
    birth_rates = [float(d) for d in d1[1:]]
    d2 = datas[2].strip().split(',')
    mortalitys = [float(d) for d in d2[1:]]
    d3 = datas[3].strip().split(',')
    growth_rates = [float(d) for d in d3[1:]]
    
#列表元素逆序
years.reverse()
birth_rates.reverse()
mortalitys.reverse()
growth_rates.reverse()

# 可以多次调用plt.plot，以便在同一个图上显示多个序列
plt.plot(years, birth_rates, 'g', label = '人口出生率') # 绿色实线
plt.plot(years, mortalitys, 'r-.', label = '人口死亡率') # 红色点虚线
plt.plot(years, growth_rates, 'b:', label = '人口自然出生率') # 蓝色点线

# 因为已经对每个序列都指派了标记，所以可以自由地布置图例，loc=9指的是“顶部中央”
plt.legend()
plt.xlabel("年度")
plt.ylabel("‰")
plt.xticks(range(2001 , 2021) , rotation = 45)
plt.title("中国人口出生率、死亡率、自然出生率变化")
plt.show()
```

### 12.2.2 绘制柱状图（条形图）

#### 例12-5，2011-2020年中国人口的柱状图

```python
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline 
#%matplotlib qt5

# 指定默认字体
mpl.rcParams['font.sans-serif'] = ['FangSong'] 
# 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['axes.unicode_minus'] = False 

with open('中国人口年度数据.csv') as f:
    datas = f.readlines()
    title = datas[0].strip().split(',') 
    years = [ int(y[:-1]) for y in title[1:11]]
    d1 = datas[1].strip().split(',')
    Populations = [float(d) for d in d1[1:11]]
       
#列表元素逆序
years.reverse()
Populations.reverse()

plt.bar(years , Populations)
plt.xlabel("年度")
plt.ylabel("人口数量（万）")
plt.title("2011-2020年中国人口数量")
#使用年度作为X轴刻度
plt.xticks([i for i in range(2011, 2021)])
plt.show()
```

#### 例12-5-1，2011-2020年中国人口男女对比的柱状图

```python
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline 
#%matplotlib qt5

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
# 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['axes.unicode_minus'] = False 

with open('中国人口年度数据.csv') as f:
    datas = f.readlines()
    title = datas[0].strip().split(',') 
    years = [ int(y[:-1]) for y in title[1:11]]
    d2 = datas[2].strip().split(',')
    M = [float(d) for d in d2[1:11]]
    d3 = datas[3].strip().split(',')
    F = [float(d) for d in d3[1:11]]
    
#列表元素逆序
years.reverse()
M.reverse()
F.reverse()

x1 = [x*2 for x in range(1 , 11)]
x2 = [x + 0.8  for x in x1]
plt.bar(x1, M, label = 'Male')
plt.bar(x2, F , label = 'Female')
plt.xlabel("年度")
plt.ylabel("人口数量（万）")
plt.title("2011-2020年中国人口数量")
#使用年度作为X轴刻度
plt.xticks([ x + 0.4  for x in x1],[i for i in range(2011, 2021)])
plt.legend(loc = 8)
plt.show()
```

#### 12-7，考试分数分布图

```python
#考试分数分布图
import numpy as np 
import matplotlib as mpl
import matplotlib.pyplot as plt 
%matplotlib inline 

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
# 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['axes.unicode_minus'] = False

grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
#定义一个lambda函数，将每个成绩转换为10分段的数据
decile = lambda grade: grade // 10 * 10
G = [decile(grade) for grade in grades]
#统计每个分数段的成绩数
counts = {}
for g in G:
    counts[g] = counts.get(g , 0) + 1

#每个条形向左侧移动4个单位,给每个条形设置正确的高度，宽度设为8
plt.bar([x - 4 for x in counts.keys()] , counts.values(),8) 
plt.axis([-5, 105, 0, 5]) # x轴取值从-5到105, y轴取值0到5
plt.xticks([10 * i for i in range(11)]) # x轴标记为0，10，...，100
plt.xlabel("分数段")
plt.ylabel("学生数")
plt.title("考试分数分布图")

plt.show()
```

#### 例12-8,绘制2020年中国人口那女比例

```python
import numpy as np 
import matplotlib as mpl 
import matplotlib.pyplot as plt 
%matplotlib inline 
#%matplotlib qt5

mpl.rcParams['font.sans-serif'] = ['FangSong'] # 指定默认字体
# 解决保存图像是负号'-'显示为方块的问题
mpl.rcParams['axes.unicode_minus'] = False 

with open('中国人口年度数据.csv') as f:
    datas = f.readlines()
    title = datas[0].strip().split(',') 
    year = int(title[1][:-1])
    d2 = datas[2].strip().split(',')
    male = float(d2[1])
    d3 = datas[3].strip().split(',')
    female = float(d3[1])

plt.pie([male, female], labels = ['男', '女'], 
        autopct='%1.1f%%',
        shadow=True, 
        startangle=90)

plt.title("2020年中国人口男女构成")

plt.show()
```
