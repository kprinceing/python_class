#### 例5.3 定义一个函数F(n)，求第n个斐波那契数列的值

```python
#定义一个函数F(n)，求第n个斐波那契数列的值
def F(n):
    if n in {1 , 2}:
        return 1
    elif n >= 3 :
        F_1 = 1
        F_2 = 1
        for i in range(3 , n + 1 ) :
            Fn = F_1 + F_2
            F_2 , F_1 = F_1 , Fn
        return Fn
```

#### 例5.5：定义一个绘制五角星的函数并调用

```python
#绘制五角星
'''
功能：绘制五角星
重点：函数的定义与调用、三角函数的使用、角度和弧度的换算'''
import turtle as t
import math

def drawStar(x, y, r, color = 'red', heading = 90):
    '绘制一个以(x,y)为圆心、以r为半径的五角星'
    t.up()
    t.goto(x, y)
    t.color(color)
    t.setheading(heading)
    edge=r * math.sin(math.radians(36))/math.sin(math.radians(126))
    t.forward(r)
    t.down()
    t.left(180-18)
    t.begin_fill()
    for i in range(5):
        t.forward(edge)
        t.right(72)
        t.forward(edge)
        t.left(144)
    t.end_fill()
```

```python
#调用drawStar()函数时缺省可选参数
drawStar(10 ,20 , 100)
t.done()
```

```python
#调用drawStar()函数时缺省可选参数
drawStar(10 ,20 , 100 , 'blue')
t.done()
```

如果此时的全局变量不是整数n，而是列表类型ls，会怎么样呢？理解如下代码。<br><br>
如果func()函数内部存在一个真实创建过且名称为ls的列表，则func()将操作该列表而不会修改全局变量，例子如下。

```python
ls = []               #ls是全局列表变量
def testfunc(a, b):
    ls.append(b)      #将局部变量b的值作为一个元素添加到全局列表变量ls中 
    return a * b

s = testfunc(5 , 2)
print(s, ls)  #测试一下ls值是否改变
```

```python
ls = []    #ls是全局列表变量
def testfunc(a, b):
    ls = []           #创建了名称为ls的局部列表变量
    ls.append(b)      #将局部变量b的值作为一个元素添局部列表变量ls中 
    return a * b

s = testfunc("Hi", 3)
print(s, ls)  #测试一下ls值是否改变
```

### 5.3.3 函数对变量的作用原则
**Python函数对变量的作用遵守如下原则：**<br>
- 简单数据类型变量无论是否与全局变量重名，仅在函数内部创建和使用，函数退出后变量被释放；<br><br>
- 简单数据类型变量在用global保留字声明后，作为全局变量；<br><br>
- 对于组合数据类型的全局变量，如果在函数内部没有被真实创建的同名变量，则函数内部可直接使用并修改全局变量的值；<br><br>
- 如果函数内部真实创建了组合数据类型变量（赋空值），无论是否有同名全局变量，函数仅对局部变量进行操作。

## 5.6 函数的递归
### 5.6.1 递归的定义

- 函数作为一种代码封装，可以被其他程序调用;<br><br>
- 函数也可以被函数自己内部代码调用。这种函数定义中调用函数自身的方式称为**递归**。<br><br>
  - 就像一个人站在装满镜子的房间中，看到的影像就是递归的结果。<br><br>
  - 递归在数学和计算机应用上非常强大，能够非常简洁的解决重要问题。
<br><br>

数学上有个经典的递归例子叫阶乘。
<br><br>
**阶乘的例子揭示了递归的2个关键特征：**<br><br>
（1）存在一个或多个**基例**，基例不需要再次递归，它是确定的表达式；<br><br>
（2）所有递归链要以一个或多个基例结尾。（因此，一般用if语句判定）

### 5.6.2 递归的使用方法

例5.10：利用递归函数进行阶乘的计算。<br><br>
根据用户输入的整数n，计算并输出n的阶乘值。

```python
#m5.1CalFactorial.py
def fact(n):
    if n == 0:
        return 1
    else:
        return n * fact(n-1)
```

```python
fact(10)
```

```python
def fact(n):
    s = 1
    for i in range(1,n+1):
        s *= i
    return s
```

```python
fact(5)
```

```python
num = eval(input("请输入一个整数: "))
print(fact(abs(int(num))))
```

### 5.6.2 递归的使用方法（续）
 
**注**：每次函数调用时，函数参数的副本会临时存储，递归中各函数再运算自己的参数，相互没有影响。当基例结束运算并返回值时，各函数逐层结束运算，想调用者返回计算结果。

#### 例11，定义一个递归函数F(n)，求第n个斐波那契数列的值

```python
#定义一个递归函数F(n)，求第n个斐波那契数列的值
def F(n):
    if n in {1 , 2}:
        return 1
    elif n >= 3 :
        return F(n-1) + F(n-2)
```

```python
F(3)
```

```python
n = int(input('请输入一个自然数n：'))
if n > 0 :
    print('F({})={}'.format(n, F(n)))
else:
    print('输入的n值错误！')
```

#### 例12，定义一个递归函数koch(size, n)，绘制n阶科赫曲线，size为边长
科赫曲线是de Rham曲线的特例。
  1. 给定线段AB，科赫曲线可以由以下步骤生成：
  2. 将线段分成三等份（AC,CD,DB）
  3. 以CD为底，向外（内外随意）画一个等边三角形DMC
  4. 将线段CD移去
  5. 分别对AC,CM,MD,DB重复1~3。

```python
#绘制科赫曲线

import turtle

def koch(size, n):      # 两个参数 size为长度大小，n为阶数
    if n==0:                            #  0阶特判，为一条直线，长度即为size
        turtle.fd(size)
    else:                               # 高阶科赫雪花曲线，有尖存在
        for i in [0, 60, -120, 60]:         # turtle在“尖”的四条线上改变的角度，分别为0°，60°，-120°，60°
            turtle.left(i)                  # 完成转弯操作
            koch(size/3, n-1)               # 递归操作，完成每个边上的下一阶绘制
```

```python
koch(300, 0)
turtle.done()
```

```python
koch(300, 1)
turtle.done()
```

```python
koch(300, 2)
turtle.done()
```

```python
koch(300, 5)
turtle.done()
```

**科赫雪花**

一个边长为1的等边三角形，取每边中间的三分之一，接上去一个形状完全相似的但边长为其三分之一的三角形，结果是一个六角形。取六角形的每个边做同样的变换，即在中间三分之一接上更小的三角形，以此重复，直至无穷。外界的变得原来越细微曲折，形状接近理想化的雪花。


```python
import turtle
def koch(size, n):
    if n == 0:
        turtle.fd(size)
    else:
        for angle in [0, 60, -120, 60]:
            turtle.left(angle)
            koch(size/3, n-1)

def koch_snowflower(n):
    turtle.setup(800,800)
    turtle.speed(0)  #控制绘制速度
    turtle.penup()
    turtle.goto(-300, 150)
    turtle.pendown()
    turtle.pensize(2)
    for i in range(3):
        koch(600,n)     # 0阶科赫曲线长度，阶数
        turtle.right(120)
    turtle.hideturtle()
    turtle.done()
```

```python
koch_snowflower(0)
```

```python
koch_snowflower(5)
```
